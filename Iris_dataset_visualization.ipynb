**FacetGrid Plot:**
It allows to create multiple plots, each showing a subset of the data, based on the values of one or more categorical variables.

```python
import matplotlib.pyplot as plt
%matplotlib inline

sns.FacetGrid(iris,hue='Species').map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

**Boxplot or Whisker plot:**
Box plot give a statical summary of the features being plotted.

```python
iris.head()
```

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Species',y='PetalLengthCm',data=iris,order=['Iris-virginica','Iris-versicolor','Iris-setosa'],linewidth=2.5,orient='v',dodge=False)
```

```python
#iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
iris.boxplot(by="Species", figsize=(12, 6))
```

**Strip plot:**
It is a categorical scatter plot where the data points are displayed as individual points along an axis. 

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray',size=8,palette='winter',orient='v')
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.title("Strip Plot of Sepal Length by Species")
plt.show()
```

**Combining Box and Strip Plots**

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray')
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

**Violin Plot:**

**LM PLot:**
It is used to visualize the linear relationship between two variables along with the associated uncertainty.

**KDE Plot:**
A KDE (Kernel Density Estimate) plot is a method used to visualize the probability density function of a continuous random variable.

```python
# Creating a kde plot of sepal_length versus sepal width for setosa species of flower.
sub=iris[iris['Species']=='Iris-setosa']
sns.kdeplot(data=sub[['SepalLengthCm','SepalWidthCm']],color="plasma", shade=True, shade_lowest=False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')
```

```python
sns.FacetGrid(iris, hue="Species").map(sns.kdeplot, "PetalLengthCm").add_legend()
plt.show() 
```

**Stacked Histogram:**
It is a type of histogram where multiple histograms are stacked on top of each other to display the distribution of each category on top of the others.

```python
iris['Species'] = iris['Species'].astype('category')
#iris.head()
```

```python
list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```

With Stacked Histogram we can see the distribution of Sepal Length of Different Species together.This shows us the range of SepalLength for the three different Species of Iris Flower.
